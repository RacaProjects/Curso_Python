```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


comparacion_ventas_exp = pd.read_csv("/content/sample_data/comparacion_ventas_exp.csv")
comparacion_mercado_exp = pd.read_csv("/content/sample_data/composicion_mercado_exp.csv")
correlacion_marketing_exp = pd.read_csv("/content/sample_data/correlacion_marketing_exp.csv")
distribucion_tiempos_exp = pd.read_csv("/content/sample_data/distribucion_tiempos_exp.csv")
```

```python
# Grafica de correlacion de las ventas generadas
#df = sns.load_dataset("correlacion_marketing_exp")

sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"], y = correlacion_marketing_exp["Ventas generadas (USD)"])
```

```python
#Cambio de color
sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"], y = correlacion_marketing_exp["Ventas generadas (USD)"], color= "red",)
```

```python
# Cambio de estilo con style = ""
sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"], y = correlacion_marketing_exp["Ventas generadas (USD)"], color= "red", style = "")

#no se pudo ejecutar por insuficientes datos
```

```python

sns.scatterplot(x=correlacion_marketing_exp["Inversión en marketing (USD)"],
                y = correlacion_marketing_exp["Ventas generadas (USD)"],
                hue = correlacion_marketing_exp ["Campaña ID"] ,
                legend = True)
```

```python
#nombre de los ejes

# Convertir la columna "Fecha" a formato datetime
x = pd.to_datetime(comparacion_ventas_exp["Fecha"])

# Extraer la columna "Ventas"
y = comparacion_ventas_exp["Ventas"]

# Crear la figura y el eje
fig, ax = plt.subplots()

# Graficar los datos en el eje
ax.plot(x, y)

# Mostrar el gráfico
plt.show()
```
